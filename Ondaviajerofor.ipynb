# Ondas Viajeras

El ciclo for nos ayuda a observar cómo se comporta cada gráfica con los diferentes valores que toma c y t.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 1, 500)

# Función de perturbación
def f(x_p):
    return np.exp(-((x_p - 0.5)**2) / 0.01)

# Valores
valoresc = [0.3, 0.5, 0.7, 1.0, 1.2]
valorest = [0.1, 0.3, 0.5, 0.7,1.0]

#Diferentes valores para c y t
for c in valoresc:
    for t in valorest:
        # Calcular las ondas
        f_derecha = f(x - c*t)
        f_izquierda = f(x + c*t)
        f_total = f_derecha + f_izquierda

        # Gráfica
        plt.figure(figsize=(10, 5))

        # Graficar cada onda
        plt.plot(x, f_derecha, label="Onda a la derecha", color="magenta")
        plt.plot(x, f_izquierda, label="Onda a la izquierda", color="lightcoral")
        plt.plot(x, f_total, label="Superposición", linestyle='--', color='darkorchid')

        plt.xlim(-0.5, 1.5)
        plt.ylim(0, 1.25)
        plt.title(f"Ondas para c = {c}, t = {t}")
        plt.xlabel("x")
        plt.ylabel("f(x')")
        plt.legend()
        plt.grid(True)

        #Graficamos
        plt.show()

```
