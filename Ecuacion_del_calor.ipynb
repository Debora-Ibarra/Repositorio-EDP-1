# Serie de Fourier de una función a trozos

Nuestra función esta definida por:


$$
f(x) =
\begin{cases}
x+2 & \text{si } -2 \leq x < 0 \\
0 & \text{si } 0 \leq x \leq 2
\end{cases}
$$



Obteniendo los términos para la serie de Fourier:
$$a_0=1$$
De igual forma:
$$a_n=\frac{2(1-(-1)^n)}{(n\pi)^2}$$
Por último:
$$b_n=\frac{-2}{n\pi}$$

Con eso, obtenemos la siguiente serie:


$$g(x)=\frac{1}{2}+\frac{2}{\pi^2}\sum_{n=1}^{\infty}(\frac{2(1-(-1)^n)}{(n\pi)^2} \cos(\frac{n\pi(x)}{2})- \frac{-2}{n\pi}  \sin(\frac{n\pi(x)}{2})).$$

A continuación graficaremos nuestra función y nuestra serie de Fourier

```python
from pylab import *
```

```python
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        if i  < len(x)/2:
          y[i] = x[i]+2
        else:
          y[i] = 0 * x[i]
    return y
```

```python
x = linspace(-2,2,1000)
f = recta(x)
plot(x,f,label='$f(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Gráfica de nuestra función definida a trozos")
xlim([-2.01,2.01]),ylim([-0.1,2.01]),grid(),legend()
show()
```

Definimos nuestra serie de Fourier y la graficamos

```python
def u_0(x,N=20): # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1,N):
        f += ((1-(-1)**n)/n**2)*cos(n*pi*x/2)-pi/n*sin(n*pi*x/2)
    return (2/pi**2)*f+1/2
```

```python
f_n = u_0(x)
plot(x,f_n,label='$f(x)$ con $N=20$')
plot(x,f,label='$g(x)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Función definida a trozos y Serie de Fourier de la misma")
xlim([-2.01,2.1]),ylim([-2.01,2.2]),grid(),legend()
show()
```

Podemos oberservar el teorema porpuesto en la siguiente gráfica, dónde se cumple que pasa por el promedio de nuestro intevalo de la función, en este caso es 0 y 2, por lo tanto observamos como todas nuestras gráficas pasan por ese punto.

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$g(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Serie de Fourier con diferentes armónicos")
xlim([-2.01,2.01]),ylim([-2.01,2.2]),grid(),legend()
show()
```
