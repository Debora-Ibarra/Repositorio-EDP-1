#Series de Fourier en Senos y Cosenos

Ejercicio 1

```python
from sympy import fourier_series, pi, plot, Piecewise
from sympy.abc import x
import sympy as sp
```

```python
f_1=Piecewise (
    (sp.exp(2*x), (x >= 0) &(x<=1)),
    (-sp.exp(-2*x),(x >= -1) & (x < 0))
    )

```

```python
plot(f_1, (x,-2,2),line_color ='red')
```

```python
s_1 = fourier_series(f_1, (x, -1, 1))
s1 = s_1.truncate(n = 3)
s2 = s_1.truncate(n = 5)
s3 = s_1.truncate(n = 7)
p = plot(f_1, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
s_1
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```
